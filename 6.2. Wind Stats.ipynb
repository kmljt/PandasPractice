```python
# GitHub Solution

pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', 
            sep="\s+", 
            parse_dates=[[0,1,2]]) 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data', sep='\s+')
data
```

```python
data[['Yr', 'Mo', 'Dy']]
```

```python
date_str_series = data.Yr.astype(str) + '-' + data.Mo.astype(str) + '-' + data.Dy.astype(str)
date_str_series
```

```python
data['datetime'] = pd.to_datetime(date_str_series, format='%y-%m-%d')
data
```

```python
data.set_index('datetime', inplace=True)
```

```python
data.drop(columns=['Yr', 'Mo', 'Dy'], inplace=True)
```

```python
data
```

### Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.

```python
data.reset_index(inplace=True)
```

```python
data.datetime = data.datetime.apply(lambda x: x - pd.DateOffset(years=100) if x.year > 2000 else x)
data
```

### Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].


### Step 11. Find the average windspeed in January for each location.  
#### Treat January 1961 and January 1962 both as January.

```python
data[data.index.month == 1].mean()
```

```python
data[data.index.month == 1 & data.index.year.isin([1961, 1962])].mean()
```

### Step 12. Downsample the record to a yearly frequency for each location.

```python
data['year'] = data.index.year
data
```

```python
data.groupby('year').mean()
```

```python
# GitHub Solution

data.groupby(data.index.to_period('Y')).mean()
```

### Step 13. Downsample the record to a monthly frequency for each location.

```python
data['month'] = data.index.month
data
```

```python
data.groupby(['year', 'month']).mean()
```

```python
# GitHub Solution

data.groupby(data.index.to_period('M')).mean()
```

### Step 14. Downsample the record to a weekly frequency for each location.

```python
weeks = np.arange(len(data)) // 7 + 1
weeks = np.insert(weeks, 0, values=0)[:-1]
weeks
```

```python
data['week'] = weeks
data.drop(columns=['year', 'month'], inplace=True)
# data = data.loc[:, ~data.columns.isin(['year', 'month'])]

data
```

```python
data.groupby('week').mean()
```

```python
# GitHub Solution

data.groupby(data.index.to_period('W')).mean()
```

### Step 15. Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
first_52_weeks = data[(data.week > 0) & (data.week <= 52)]
first_52_weeks
```
