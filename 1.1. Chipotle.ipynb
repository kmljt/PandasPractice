### Step 9. Which was the most-ordered item? 

```python
chipo.groupby('item_name').sum()['quantity'].sort_values().tail(1)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
chipo.groupby('item_name').sum()['quantity'].max()
```

### Step 11. What was the most ordered item in the choice_description column?

```python
chipo.groupby('choice_description').sum()['quantity'].sort_values().tail(1)
```

### Step 12. How many items were ordered in total?

```python
chipo.quantity.sum()
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
chipo.dtypes, chipo['item_price']
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
a = lambda x: float(x[1:-1])
chipo.item_price = chipo.item_price.apply(a)

chipo
```

#### Step 13.c. Check the item price type

```python
chipo.item_price.dtype
```

### Step 14. How much was the revenue for the period in the dataset?

```python
rev = (chipo['quantity'] * chipo['item_price']).sum()

rev
```

### Step 15. How many orders were made in the period?

```python
# chipo.order_id.value_counts().count()
n_orders = chipo.order_id.unique().__len__()
# chipo.tail(1).order_id

n_orders
```

### Step 16. What is the average revenue amount per order?

```python
# Solution 1

rev / n_orders
```

```python
# Solution 2

chipo['revenue'] = chipo.quantity * chipo.item_price

chipo
```

```python
chipo.groupby('order_id').revenue.sum().mean()
```
